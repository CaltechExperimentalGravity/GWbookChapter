# Plots of Rates and Sensitivities as a function of Time

```python
# Library Imports and Python parameter settings
%matplotlib inline
#import sympy as sym
import numpy as np
from __future__ import division
import matplotlib.pyplot as plt
#import scipy.signal as sig
import scipy.constants as const
#from IPython.core.display import Image

# Update the matplotlib configuration parameters:
plt.rcParams.update({'font.size': 18, 
                     'font.family': 'serif', 
                     'font.weight': 'bold',
                     'legend.numpoints': 1})
```

```python
# Evolution of GW source strengths and rates

```

```python
# Plot sensitivity evolution history

#   Year   strain/rHz  BNSrange
b = np.array([
    [1965,  1e-14,  1e-7],   # Weber    1970
    [1998,  3e-21,  1e-7]    # Allegro PRD 1999
    ])

h = np.array([
    [1971,  1e-15,  1e-7],   # Forward  1971  (dx ~ 3e-16, L ~0.2-0.3 m, BW ~ 1 Hz)
    [1978,  1e-16,  1e-6],   # Forward  1978
    [1988,  1e-19,  0.25e-3],   # Garching 30m 1988 (PRD); BNS range estimated / not calculated
    [1994,  1e-20,  25e-3],   # Caltech 40m - Nov. 1994; PRL 1999
    [2001,  7e-21,  28e-3],   # TAMA  - Sep 2001; CQG 2002
    [2002,  5e-21 ,  0.1],    # L1 S1 - Aug 2002
    [2003,  3e-22 ,    1],    # L1 S2 - Mar 2003
    [2006,  3e-23 ,   15],    # H1 S5
    [2010,  1.5e-23,  20],    # L1 S6
    [2015,  0.8e-23,  60]     # L1 March 2015
    ])

fu = np.array([
    [2017,  3e-24, 205],   # aLIGO
    [2025,  1e-24, 900],   # LIGO 3
    [2035,  4e-25, 1400]   # ET
     ])
```

```python
fig = plt.figure(figsize=(12, 9))

ax1 = fig.add_subplot(111)
ax1.semilogy(b[:,0] ,  b[:,1], 'rs', alpha=0.65, markersize=13)
ax1.semilogy(h[:,0] ,  h[:,1], 'bp', alpha=0.65, markersize=13)
ax1.semilogy(fu[:,0], fu[:,1], 'gs', alpha=0.65, markersize=16)

plt.xlabel(r'Time [y]', fontsize=18)
plt.ylabel(r'Strain [1/$\sqrt{\rm Hz}$]', fontsize=18)

plt.title("Sensitivity Evolution of GW Detectors");
plt.xlim((1960, 2040));
plt.ylim((1e-25, 3e-14));
plt.legend(('Resonant Bars', 'Past Interferometers', 'Future Interferometers'),
          fancybox=True, loc='center left', fontsize='12')

# now, the second axes that shares the x-axis with the ax1
ax2 = fig.add_subplot(111, sharex=ax1, frameon=False)
ax2.semilogy(b[:,0] ,  b[:,2], 'ro', alpha=0.65, markersize=13)
ax2.semilogy(h[:,0] ,  h[:,2], 'm*', alpha=0.65, markersize=18)
ax2.semilogy(fu[:,0], fu[:,2], 'gd', alpha=0.65, markersize=16)
plt.grid(color='gray', alpha=0.15, linestyle='dashed', linewidth=4)
ax2.yaxis.tick_right()
ax2.yaxis.set_label_position("right")
plt.ylabel("Binary NS Range [Mpc]")


plt.legend(('Resonant Bars', 'Past Interferometers', 'Future Interferometers'),
          fancybox=True, loc='center right', fontsize='14')

plt.savefig("RateSensEvo.pdf")

plt.show()
```
